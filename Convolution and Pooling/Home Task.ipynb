```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import tensorflow as tf
```

```python
image = Image.open('sample_image.jpg').convert('L')
gray_image = np.array(image)

gray_image_float32 = gray_image.astype(np.float32)
```

```python
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
kernel = np.array([[1, 0, -1],
                   [1, 0, -1],
                   [1, 0, -1]])
```

```python
kernel_tensor = tf.constant(kernel, dtype=tf.float32)
kernel_tensor = tf.expand_dims(tf.expand_dims(kernel_tensor, axis=-1), axis=-1)

stride = 4
convolved_image = tf.nn.conv2d(np.expand_dims(np.expand_dims(gray_image, axis=0), axis=-1), kernel_tensor, strides=[1, stride, stride, 1], padding='SAME')
convolved_image = np.squeeze(convolved_image)
```

```python
plt.imshow(convolved_image, cmap='gray')
plt.title('Convolved Image')
plt.axis('off')
plt.show()
```
