```python
import pandas as pd
```

# Read salaries.csv using pandas and show the first five rows of dataset


```python
salaries = pd.read_csv("salaries.csv")
salaries.head()
```

```python
salaries = salaries.drop("Unnamed: 4", axis = 1)
salaries = salaries.drop("Unnamed: 5", axis = 1)

salaries = salaries.dropna()
salaries
```

# Save the salary_more_then_100k column to different variable

```python
y = salaries["salary_more_then_100k"].copy()
y
```

# Drop the salary_more_then_100k Column from Dataset

```python
#Because the Last Column is our Target Variable so we will drop this column and save that saperatly
salaries = salaries.drop("salary_more_then_100k", axis = 1)
salaries
```

# Labels to numbers 

```python
#Ml Model does not work on labels so we need to convert the labels to numbers.
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()  ## We create the label encoder for each column
#Create Label Encoder for other two columns as well
"Your Code Here"
```

```python
salaries['company_n'] = le_company.fit_transform(salaries['company']) # It will transform the company column to new column company_n
# You have to transform other column 

salaries['job_n'] = le_company.fit_transform(salaries['job'])
salaries['degree_n'] = le_company.fit_transform(salaries['degree'])
```

```python
#Show the updated Inputs
```

# Drop the Company job and Degree Column

```python
salaries = salaries.drop("company", axis = 1)
salaries = salaries.drop("job", axis = 1)
salaries = salaries.drop("degree", axis = 1)

salaries
#show the updated Inputs
```

```python
y
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(salaries, y)
```

```python
model.score(salaries, y)
```

**Is salary of Google, Computer Engineer, Bachelors degree > 100 k ?**

```python
test_row = pd.DataFrame({
    "company_n": [2],
    "job_n": [1],
    "degree_n": [0],
})
print(model.predict(test_row))
```

**Is salary of Google, Computer Engineer, Masters degree > 100 k ?**

```python
test_row = pd.DataFrame({
    "company_n": [2],
    "job_n": [1],
    "degree_n": [1],
})
print(model.predict(test_row))
```
